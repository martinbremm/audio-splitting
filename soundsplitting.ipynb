```python
import pandas as pd
df = pd.read_csv("./prodigy-v1.csv")
df = df[["audio_spans","path"]]
df["filename"] = [path[12:-4] for path in df["path"]]
```

```python
folders = []
subfolders = []
for f in df["filename"]:
    folder = f[f.find("DIGI"):f.find("DIGI")+5] 
    folder = folder[0:4]+"-"+folder[4:] # adapting to new folder format
    folders.append(folder)
    
    subfolder = f[f.find("DIGI"):f.find("-")]
    subfolders.append(subfolder)
    
df["folder"] = folders
df["subfolder"] = subfolders
```

```python
from pydub import AudioSegment
import os

out_dir = "output"
if not os.path.exists(out_dir):
    os.makedirs(out_dir)
    
    
for f in range(0,len(df)):
    filename = df["filename"][f]
    folder = df["folder"][f]
    subfolder = df["subfolder"][f]
    old_path = os.path.join(folder, subfolder, filename+".wav")
    try:
        full_clip = AudioSegment.from_wav(old_path) # loading audio file from folder
        clips = df["audio_spans"][f]
        for i, c in enumerate(clips):
            start = c["start"]*1000
            end = c["end"]*1000
            split_clip = full_clip[start:end] # splicing based on the prodigy start- and end-points
            
            new_path = os.path.join(out_dir, folder, subfolder)
            if not os.path.exists(new_path):
                os.makedirs(new_path)
            new_filename = f"{filename}-split{i}.wav"
    
            split_clip.export(os.path.join(new_path, new_filename), format="wav") # creating splits with numerated names
    except FileNotFoundError:
        print(f"{filename} not found")
```
