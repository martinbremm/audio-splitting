```python
import pandas as pd
df = pd.read_csv("./prodigy-v1.csv")
df = df[["audio_spans","path"]]
df["filename"] = [path[12:-4] for path in df["path"]]
```

```python
folders = []
subfolders = []
for f in df["filename"]:
    folders.append(f[f.find("DIGI"):f.find("DIGI")+5])
    subfolders.append(f[f.find("DIGI"):f.find("-")])
df["folder"] = folders
df["subfolder"] = subfolders
```

```python
from pydub import AudioSegment
import os

filenames = df["filename"]

log_dir = "logs/test/"
if not os.path.exists(log_dir):
    os.makedirs(log_dir)
    
    
for f in range(0,len(df)):
    filename = df["filename"][f]
    old_dir = df["subfolder"][f]
    old_path = os.path.join(old_dir, filename+".wav")
    try:
        full_clip = AudioSegment.from_wav(old_path) # loading audio file from folder
        clips = df["audio_spans"][f]
        for i, c in enumerate(clips):
            start = c["start"]*1000
            end = c["end"]*1000
            split_clip = full_clip[start:end] # splicing based on the prodigy start- and end-points
            
            new_filename = os.path.join(log_dir, f"{filename}-split{i}.wav")
            split_clip.export(new_filename, format="wav") # creating splits with numerated names
    except FileNotFoundError:
        print(f"{filename} not found")
```
